```
from scipy import integrate
import numpy as np 
import matplotlib.pyplot as plt

def f(yf, t):
  v = yf[0]
  y = yf[1]
  f0 = -y + np.sin(t) - np.cos(2*t)
  f1 = v
  return [f0, f1]
```

```
#массив точек интегрирования
ti = np.linspace(0, 7*np.pi, 100)
#начальная координата 
x0 = np.pi
#начальная скорость 
v0 = 2.0
#список начальных условий
y0 = [x0, v0]
```

```
#рещение ОДУ
sol = integrate.odeint(f, y0, ti)
#Получение гармоник решения системы ОДУ
y1=sol[:,0]  #первая производная
y2=sol[:,1]  #искомая функция
```

```
#Формирование обьекта, рисунок
fig = plt.figure(facecolor='white')
ax = fig.add_subplot(111)
ax.set(title='Решение ОДУ 2-го порядка')
ax.set_xlabel('t')
ax.set_ylabel('f')
ax.plot(ti, y1, color ='grey', marker = '*')
ax.plot(ti, y2, 'g.')
lgnd = ax.legend(['y1', 'y2'], shadow = True)
lgnd.get_frame().set_facecolor('#ffb23a')
plt.grid(True)
plt.show()
```
