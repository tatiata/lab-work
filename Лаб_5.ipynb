```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
StateNames=pd.read_csv('StateNames.csv')
```

```python
StateNames
```

```python
StateNames.info()
```

# 1.1. Box plots

```python
sns.boxplot(StateNames['Year'])
```

**Kernel** **density** **estimate**

```python
sns.kdeplot(StateNames['Year'])
```

# 1.2. Violin plot

```python
sns.violinplot(x='State', y='Year', data=StateNames)
```

```python
sns.violinplot(x='Gender', y='Year', data=StateNames)
```

# 1.3. Count plots

```python
sns.countplot(StateNames['State'])
```

# 1.4. Heatmap

```python
sns.heatmap(StateNames.corr(),annot=True,cmap='YlGnBu')
```

# 1.5. Facet grid

```python
g = sns.FacetGrid(StateNames, col='Gender', row='State')
g.map(plt.hist, 'Year')
```

# 1.6. Regplot

```python
sns.regplot(x='Count', y='Year', data=StateNames  )
```

# lmplot

```python
sns.lmplot(x='Year', y='Count', data=StateNames, col='Gender')
```

# swarm plot

```python
sns.swarmplot(x='Gender', y='Year', data=StateNames)
```

```python
sns.swarmplot(x='Count', y='Year', data=StateNames)
```

# 1.9. Pair plot

```python
sns.pairplot(data=StateNames)
```

**В** **цвете**

```python
#StateNames
sns.pairplot(StateNames, hue='Gender', palette='Set1')
```

# 1.10. Joint plot

```python
sns.jointplot(x='Gender', y='Year', data=StateNames)
```
