#  Matplotlib Basic
2.1.6.


```
import matplotlib.pyplot as plt
x = [10, 20, 30]
y = [40, 10, 30]
x1 = [10, 20, 30]
y1 = [20, 40, 10]

fig, ax = plt.subplots()

ax.plot(x, y, 'r', linewidth = 5, label = 'line1, width-5 ')
ax.plot(x1, y1, 'b',linewidth = 3, label = 'line2, width-3')
ax.set_title("Две линии с различной толщеной линии и легендами")
ax.set_xlabel('ось Х')
ax.set_ylabel('ось Y')
ax.legend()

plt.show()
```

# Matplotlib Bar
2.2.7.

```
import matplotlib.pyplot as plt
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x_pos = [i for i, _ in enumerate(x)]
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("PopularitY of Programming Language\n" + "Worldwide, Oct 2017 compared to a year ago")
plt.xticks(x_pos, x)
# Select the position of each barplots on the x-axis (space=1,3,3,2,1)
y_pos = [0,1,4,7,9,10]
# Create bars
plt.bar(y_pos, popularity)
# Turn on the grid
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

# Matplotlib - круговая диаграмм

2.3.3.

```
import matplotlib.pyplot as plt
import numpy as np 

fig=plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')

langs = ['Java', 'Python', 'PHP', 'JavaScript', 'C', 'C++']
students = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
explode = [0.1,0,0,0,0,0.1]
ax.pie(students, labels = langs, explode=explode, shadow=True, wedgeprops={'lw':1, 'ls':'-','edgecolor':"k"}, autopct='%1.2f%%')

ax.set_title("Popularity of Programming Language Worldwide, Oct 2017")

plt.show()
```
