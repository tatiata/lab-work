# Задание 4 

```python
import pandas as pd
import numpy as np
```

```python
from ast import literal_eval as le 
df1, df2 = map(le,input().split('-'))
```

```python
df1 = pd.DataFrame(df1)
df2 = pd.DataFrame(df2)

df1['prod_id'] = np.arange(1,26)
```

```python
df = df1.merge(df2, how='outer')
df
```

```python
df0 = df.fillna(0)
df0
```

```python
df3=df0.groupby(['availability']).get_group('yes')
df3
```

```python
df4=df3.groupby(['type']).mean()
df4
```

```python
df5=df4.drop('prod_id', 1)
df5
```

```python
df5.columns.name = 'Prod_market'
df5
```
